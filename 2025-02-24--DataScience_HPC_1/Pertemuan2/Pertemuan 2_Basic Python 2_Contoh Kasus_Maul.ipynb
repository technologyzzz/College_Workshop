<h3 id="comp">Operator Perbandingan</h3>

Operasi perbandingan membandingkan beberapa nilai atau operan dan berdasarkan suatu kondisi, menghasilkan Boolean. Saat membandingkan dua nilai, dapat digunakan operator sebagai berikut:

<ul>
     <li>sama: <b>==</b></li>
     <li>tidak sama: <b>!=</b></li>
     <li>lebih besar dari: <b>></b></li>
     <li>kurang dari: <b>&lt;</b></li>
     <li>lebih besar dari atau sama dengan: <b>>=</b></li>
     <li>kurang dari atau sama dengan: <b>&lt;=</b></li>
</ul>

Berikan variabel <code>a</code> nilai 5. Gunakan operator persamaan, yang dilambangkan dengan dua tanda sama dengan (==), untuk menentukan apakah dua nilai sama. Contoh berikut membandingkan variabel <code>a</code> dengan 6.

````python
# Operator Perbandingan: Apakah dua nilai sama?

a = 5
a == 6
````

Hasilnya adalah <b>False</b>, karena 5 tidak sama dengan 6.

Pertimbangkan operator perbandingan kesetaraan berikut: <code>i > 5</code>. Jika nilai operan kiri, dalam hal ini variabel <b>i</b>, lebih besar dari nilai operan kanan, dalam hal ini 5, maka pernyataannya adalah <b>True</b>. Jika tidak, pernyataannya adalah <b>False</b>. Jika <b>i</b> sama dengan 6, karena 6 lebih besar dari 5, outputnya adalah <b>True</b>.

````python
# Tanda lebih besar dari (>)

i = 6
i > 5
````

Deklarasikan <code>i = 2</code>. Pernyataan <b>False</b> karena 2 tidak lebih besar dari 5:

````python
# Tanda lebih besar dari (>)

i = 2
i > 5
````

### Operator Logika

Dalam beberapa kasus, diperlukan pemeriksaan lebih dari satu kondisi secara bersamaan. Misalnya, dapat diperiksa **apakah dua kondisi sekaligus bernilai True.** Operator logika digunakan untuk menggabungkan atau memodifikasi kondisi tersebut.

Beberapa operator logika yang umum digunakan antara lain:
* <code>AND</code>
* <code>OR</code>
* <code>NOT</code>

Operator-operator ini dapat diringkas dalam bentuk tabel kebenaran untuk dua variabel, sebagai berikut:

image.png

Pernyataan <code>AND</code> hanya **True** jika kedua kondisi benar. Pernyataan <code>OR</code> bernilai True jika satu kondisi, atau keduanya **True**. Pernyataan <code>NOT</code> menghasilkan nilai kebenaran yang berlawanan.

#### Contoh Kasus:
Pertimbangkan cara menentukan apakah sebuah album dirilis setelah tahun 1979 (tidak termasuk 1979) dan sebelum tahun 1990 (tidak termasuk 1990). Dengan demikian, periode antara tahun 1980 hingga 1989 memenuhi kondisi tersebut.

Hal ini digambarkan pada gambar di bawah ini. Warna hijau pada baris **a** dan **b** menunjukkan periode di mana masing-masing pernyataan bernilai **True**. Garis hijau pada **c** menandakan area di mana kedua kondisi bernilai **True**, sesuai dengan tumpang tindih wilayah hijau tersebut.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsEgTwo.png" width="650" />


````python
# Inisialisasi variabel album_year dengan tahun rilis album
album_year = 1981

# Cek apakah album dirilis setelah 1979 dan sebelum 1990
if (album_year > 1979) and (album_year < 1990):
    # Jika kondisi terpenuhi, cetak bahwa tahun album berada di antara 1980 dan 1989
    print("Tahun album antara 1980 dan 1989")

# Menandai bahwa proses pemeriksaan selesai
print("")
print("Proses Selesai")
````

### 1. Kondisi IF dengan Logika Tunggal
Seperti bahasa pemrograman lainnya, Python juga memiliki pernyataan percabangan IF. Dalam Python, ekspresi ditempatkan setelah if, dan keputusan ditentukan berdasarkan nilai kebenaran ekspresi.

Jika ekspresi bernilai <code>True</code>, blok pernyataan di dalam pernyataan if akan dieksekusi. Dengan konvensi, blok ini menjorok setelah titik dua (:). Jika ekspresi bernilai <code>False</code>, blok berikutnya (setelah pernyataan IF) dieksekusi.

Logika tunggal adalah struktur logis yang hanya memiliki satu kondisi untuk dipilih. Struktur logika tunggal menggunakan if, memiliki sintaks berikut:

```python
if(ekspresi):
     penyataan
```

Kondisi IF di bagian ini digunakan untuk menjalankan perintah jika kondisinya <code>True</code>. Sebaliknya, jika kondisinya <code>False</code>, perintah tidak akan dijalankan.

image.png


Berdasarkan flowchart di atas, menunjukkan pernyataan logis bahwa jika Skor lebih besar dari 60, maka seorang siswa dinyatakan "Lulus". Jika tidak, tidak akan terjadi apa-apa.

````python
# Masukkan skor ujian
skor = float(input("Masukkan skor ujian: "))

# Pemeriksaan kelulusan
if (skor > 60):
    print("Selamat, Anda lulus ujian!")
````

### 2. Kondisi IF–ELSE

Dalam perkembangannya, struktur logika tunggal yang menggunakan **IF** dapat dilengkapi dengan kondisi yang memiliki dua cabang hasil pemrosesan. Dengan demikian, pengambilan keputusan tidak hanya digunakan untuk menentukan tindakan yang akan diambil ketika kondisi terpenuhi, tetapi juga untuk menentukan tindakan yang akan diambil ketika kondisi tersebut tidak terpenuhi. Untuk mengantisipasi kebutuhan tersebut, sintaks **IF** dapat dikembangkan menjadi:

```python
if (expression):
    statement
else:
    statement
```

image.png


````python
# Masukkan skor ujian
skor = float(input("Masukkan skor ujian: "))

# Pemeriksaan kelulusan
if (skor > 60):
    print("Selamat, Anda lulus ujian!")
else:
    print("Maaf, Anda tidak lulus ujian.")
````

<h2 id ='loop'>Struktur Perulangan (Looping)</h2>

Struktur loop digunakan ketika ingin mengeksekusi blok perintah yang dilakukan secara berulang kali.

image.png

Flowchart di atas menunjukkan alur dari struktur loop. Suatu proses akan terus berulang selama kondisi pada ekspresi bernilai <code>True</code>. Ketika kondisi tersebut bernilai <code>False</code>, proses perulangan akan berhenti. Dalam pemrograman Python, terdapat tiga jenis perulangan yang umum digunakan, yaitu <b>while</b>, <b>for</b>, dan <b>nested loop</b>.

### 1. Perulangan dengan While (While Loop)


***Loop while*** digunakan untuk mengulang selama ekspresi kontrol <code>True</code>. Diagram alir terlihat seperti di atas. Pemrosesan loop berhenti ketika ekspresi kontrol <code>False</code> atau jeda eksplisit diberikan dalam kode program. Perintah while memiliki bentuk dasar sebagai berikut:

```python
while (expression):
    statement
    statement
    statement
...
```

Sebuah <code>expression</code> adalah kondisi yang harus dipenuhi sebagai syarat terjadinya loop. Artinya, jika kondisi dalam ekspresi masih terpenuhi, maka loop akan tetap dieksekusi untuk diproses.

Blok pernyataan yang mengikuti kondisi while dan memiliki posisi <b>indentasi</b> yang sama, dianggap sebagai blok pernyataan yang akan dieksekusi.

````python
hitung = 1

while (hitung <9):
     print('Perulangan -', hitung)
     hitung = hitung + 1
print('Perulangan Selesai')
````

Berikut adalah contoh iterasi iterasi menggunakan sebuah *list* <code>dates</code> dan berhenti pada tahun 1973, kemudian mencetak jumlah iterasi. Ini dapat dilakukan dengan blok kode berikut:

````python
# Contoh Perulangan Sementara

tanggal = [1982, 1980, 1973, 2000]

i = 0
tahun = tanggal[0]

while (tahun != 1973):
     print(tahun)
     i = i + 1
     tahun = tanggal[i]

print("Butuh", i ,"kali pengulangan untuk keluar dari perulangan.")
````

### 2. Perulangan dengan For (For Loop)

Perintah <code>For</code> digunakan untuk mengulangi beberapa kali yang telah ditentukan. Loop didefinisikan dalam daftar yang telah ditentukan. Perintah for loop memiliki bentuk dasar sebagai berikut:

```python
for <target> in <object>:
    statement
```

Pada bentuk umum di atas, for memiliki 2 data yaitu <b>Target</b> dan <b>Object</b>.
- <b>Target</b>, merupakan variabel yang akan menampung setiap iterasi objek.
- <b>Object</b>, berisi list sebagai kondisi yang menentukan loop.

Contoh:

Jika <b>Object</b> berisi list angka ganjil dari 1 sampai 10, maka <b>Target</b> akan mengambil dan menampilkan elemen-elemen dalam list tersebut satu per satu.

````python
the_list = [1, 3, 5, 7, 9]
for x in the_list:
    print(x)
````

Selain itu, ***for loop*** juga dapat menggunakan fungsi <code>range</code> dengan struktur sebagai berikut:

````python
x = range(3)

for n in x:
  print(n)
````

<h3 id="rentang"><b>Range</b></h3>

Terkadang, pemrogram perlu mengulangi operasi tertentu berkali-kali. Eksekusi berulang seperti ini dilakukan oleh <b>loop</b>. Dua jenis perulangan yang umum digunakan adalah perulangan <code>for</code> dan perulangan <code>while</code>.

Sebelum membahas loop, perlu dipahami objek <code>range</code>. Range sangat berguna dan dapat dipahami sebagai list yang diurutkan. Untuk saat ini, perhatikan kasus paling sederhana. Jika ingin membuat objek yang berisi elemen berurutan dari 0 hingga 2, cukup menggunakan perintah berikut:

```python
for <target> in range(startValue, endValue, steps):
    statement
```

Contoh sintaks untuk menggunakan rentang pada <code>loop</code> menggunakan <code>for</code> adalah sebagai berikut:

````python
for x in range (5, 12, 2):
    print("Perulangan -", x)
````

### 3. Perulangan dengan Nested Loop

Nested *Loop* adalah *loop* di dalam *loop*. Perintah nested loop memiliki bentuk dasar sebagai berikut:

Loop dalam skrip berarti loop dimulai dari nilai awal 5 hingga nilai akhir 12 dengan langkah (angka bertambah) 2. Dengan demikian, output loop seperti yang terlihat pada program di atas.

image.png


```python
# Nested Loop with For Loop
for <target> in <object>:
    for <target> in <object>:
        statement
    statement

# Nested Loop with While Loop
while (expression):
    while (expression):
        statement
    statement

```

````python
for i in range(2):
    for j in range(6,10,1):
        print("i value =", i, " j =", j)
    print()
````

#### 5. Hitung Diskon:
Buat program yang meminta pengguna memasukkan total belanja. Program kemudian memberikan diskon berdasarkan aturan berikut:

- Jika total belanja di atas 100.000, diskon 10%.
- Jika total belanja di atas 50.000, diskon 5%.
- Jika total belanja di atas 25.000, diskon 2.5%.
- Jika total belanja di bawah atau sama dengan 25.000, tidak ada diskon.

````python
total_belanja = float(input("Masukkan total belanja: "))

if total_belanja > 100000:
    diskon = 0.1
elif total_belanja > 50000:
    diskon = 0.05
elif total_belanja > 25000:
    diskon = 0.025
else:
    diskon = 0

total_diskon = total_belanja * diskon
total_bayar = total_belanja - total_diskon

print("Total bayar setelah diskon:", total_bayar)
````
