# Lingkup Sebuah Variable
<hr>


Cakupan variabel dapat berupa variabel <b>lokal</b> atau <b>global</b>. Sehingga tidak semua variabel bisa diakses dari semua tempat, bergantung bagaimana cara mendefinisikan variabel tersebut.

### a) Variabel Lokal

Variabel lokal adalah variabel yang hanya dapat diakses dalam suatu fungsi, di mana variabel tersebut didefinisikan

### b) Variabel Global

Variabel global didefinisikan di luar fungsi sehingga memiliki cakupan global dan dapat diakses dari mana saja.

Berikut adalah contoh script yang akan membedakan antara variabel lokal dan variabel global:

```python
# Membuat Variabel Global
name = 'Nafa'
address = 'Palembang'

def testing():
    # Membuat Variabel Lokal
    name = 'Fankuy'
    address = 'Depok'

    # Mengakses Variabel Lokal
    print('Name : %s' %name)
    print('Address : %s' %address)

# Mengakses Variabel Global
<...>
<...>
<...>

print('-------------------------')

# Memanggil Fungsi Pengujian
print('Local Variables')
testing()
```

Pada script di atas, variabel **name** dan **address** yang ditulis pada baris pertama merupakan variabel global yang dapat dipanggil dari mana saja. Sedangkan variabel <b>name</b> dan <b>address</b> yang ditempatkan pada fungsi <b>testing()</b> adalah variabel lokal dan hanya dapat dipanggil di dalam fungsi tersebut.

# Object dan Class
<hr>

*Class* adalah sebuah *blueprint* atau prototipe yang ditentukan pengguna berdasarkan objek yang dibuat. Kelas menyediakan sarana untuk menggabungkan data dan fungsionalitas bersama-sama. Membuat kelas baru akan membuat tipe objek baru, memungkinkan <b>instance</b> baru dari tipe tersebut dibuat. Setiap instance kelas dapat memiliki <b>atribut</b> yang melekat padanya untuk mempertahankan statusnya. Instance kelas juga dapat memiliki metode (didefinisikan oleh kelasnya) untuk memodifikasi statusnya.

Misal sebuah 'Kucing' untuk memperdekat pemahaman:

- “Kucing” merupakan sebuah **definisi objek**, ia memiliki 2 telinga, 4 kaki, 1 ekor, nama dan lain-lain (sebagai **atribut**), ia juga bisa berlari, mengeong, makan dan minum (sebagai **perilaku**).

- Misalkan kita memiliki 4 kucing: berarti kita memiliki "4 **instance**" dari kelas kucing.

- Masing-masing dari 4 kucing tersebut memiliki **atribut** yang telah didefinisikan sebelumnya seperti <code>kaki, telinga, ekor, dan lain-lain.</code> Ia juga memiliki kemampuan berlari, mengeong, dan sebagainya seperti yang telah didefinisikan sebelumnya.

## Atribut and *Behavior*

- Atribut dalam OOP mewakili **variabel** yang dimiliki oleh suatu objek.
- Sedangkan perilaku (*behaviour*) dalam OOP merepresentasikan **fungsi** suatu objek.


## Contoh Atribut

Dimisalkan terdapat 3 kucing, masing-masing kucing memiliki **warna** dan **usia** sendiri.

Jika kita ingin merepresentasikan 3 kucing dengan pendekatan OOP, kita dapat menuliskan kode programnya sebagai berikut:

```python
# Kelas Kucing sebagai "definisi"
class Kucing:
  warna = None
  usia = None

# Membangun instance atau variabel sebagai "objek nyata"
kucing1 = Kucing()
kucing1.warna = "hitam"
kucing1.usia = "3 bulan"

kucing2 = Kucing()
kucing2.warna = "putih"
kucing2.usia = "2 bulan"

kucing3 = Kucing()
kucing3.warna = "kuning"
kucing3.usia = "3.5 bulan"
```

Setelah membuat **kelas**, kita bisa membuat instance sesuka kita.

Dalam contoh berikut ini, terdapat 2 buah mahasiswa: yang pertama mahasiswa dengan nama "Deni" dari Jawa Timur, dan yang kedua adalah "Lendis Fabri" yang juga dari Jawa Timur, kemudian kita akan membuat **fungsi** untuk memperkenalkan diri.
