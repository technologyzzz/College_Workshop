<h2 id='statistic'>Operasi Statistik Dasar dengan Python</h2>

Berikut ini akan diberikan contoh berbagai jenis operasi statistik dasar pada matriks menggunakan library NumPy, seperti mean, median, standar deviasi, nilai minimum maksimum dan sebagainya.

<h3 id='mean'>1) Mean (Rata-rata)</h3>

Mean adalah rata-rata atau nilai yang paling umum dalam kumpulan angka.

Dalam statistik, mean merupakan ukuran tendensi sentral dari distribusi probabilitas sepanjang median dan modus. Mean juga disebut sebagai nilai yang diharapkan.

```python
import numpy as np
val = np.array([7,6,8,9,5]) # Deklarasi Array
np.mean(val) # Fungsi mean() pada Library Numpy
```

<h3 id='median'>2) Median (Nilai tengah)</h3>

Median, dalam statistik, adalah nilai tengah dari daftar data yang diberikan, bila disusun secara berurutan. Susunan data atau pengamatan dapat dilakukan baik secara ascending maupun descending.

**Contoh**: Median dari 2, 3, 4 adalah 3.

Dalam Matematika, median juga merupakan jenis rata-rata, yang digunakan untuk menemukan nilai pusat. Oleh karena itu, disebut juga ukuran tendensi sentral.

Berikut adalah kode untuk menghitung rata-rata nilai array.

image.png

```python
val = np.array([7,6,8,9,5]) # Deklarasi Array
np.median(val) # Fungsi median() pada Library Numpy
```

<h3 id='std'>3) Standar Deviasi (Simpangan Baku)</h3>

Nilai simpangan baku adalah nilai yang digunakan dalam menentukan sebaran data dalam suatu sampel dan melihat seberapa dekat data tersebut dengan nilai rata-ratanya. Standar deviasi atau standar deviasi adalah ukuran penyebaran yang paling baik, karena menggambarkan besarnya penyebaran setiap unit pengamatan.

Berikut adalah kode untuk menghitung standar deviasi dari nilai array.

```python
val = np.array([7,6,8,9,5]) # Deklarasi Array
np.std(val) # Fungsi standar deviasi pada Library Numpy
```

<h3 id='maxmin'>4) Nilai Maksimum dan Minimum</h3>

```python
val = np.array([7,6,8,9,5,3,2,8,4,1]) # Deklarasi Array
np.min(val) # Fungsi nilai minimum pada Library Numpy
np.max(val) # Fungsi nilai maximum pada Library Numpy
```
