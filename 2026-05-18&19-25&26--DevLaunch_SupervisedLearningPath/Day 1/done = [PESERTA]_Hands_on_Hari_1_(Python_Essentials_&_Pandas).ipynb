## 2. Struktur Data (List & Dictionary)

📖 **Konsep**
* **Untuk Awam:** Jika data kita banyak, kotak sepatu tidak cukup. Kita butuh "lemari".
  * **List:** Seperti **Laci Baju Bersusun**. Kita mencari barang berdasarkan urutan nomor laci (Dimulai dari laci nomor 0).
  * **Dictionary:** Seperti **Lemari Arsip Berlabel**. Kita mencari dokumen langsung dengan menyebutkan nama map/labelnya, tidak perlu menghafal nomor urut.
* **Untuk Pro:** *List* adalah array yang *mutable* (bisa diubah isinya) dan menjaga urutan (ordered). *Dictionary* adalah implementasi dari *Hash Table* (Key-Value pair) yang proses pencariannya (*lookup time*) jauh lebih cepat ($O(1)$) dibandingkan *List* ($O(n)$) untuk dataset besar.

🎨 **Visualisasi Konsep**
```text
📂 LIST (Urutan/Indeks)                🗂️ DICTIONARY (Label/Key-Value)
[0] ---> "Budi"                        ["nama"]  ---> "Budi"
[1] ---> "Andi"                        ["divisi"]---> "Data Science"
[2] ---> "Siti"                        ["gaji"]  ---> 8000000

image.png

````
# 💻 SINTAKSIS & IMPLEMENTASI DASAR (LIST)
# List diapit oleh kurung siku []
daftar_umur = [22, 38, 26, 35, 20]

# Memanggil elemen menggunakan indeks (INGAT: Komputer mulai menghitung dari 0)
print("Umur urutan pertama : ", daftar_umur[0])
print("Umur urutan ketiga : ", daftar_umur[2])
print("Umur urutan terakhir :", daftar_umur[-1])
````

## 3. Control Flow (Memberi "Otak" pada Program)

📖 **Konsep**
* **Untuk Awam:** Sejauh ini, komputer kita cuma bisa menyimpan data, tapi belum bisa berpikir. Nah, *Control flow* ini adalah cara kita memberikan "otak" ke komputer.
  * **If-Else (Percabangan):** Ibarat Anda di jalan persimpangan. *Jika* macet, lewat tol. *Jika tidak*, lewat jalan biasa.
  * **For Loops (Perulangan):** Ibarat roda yang berputar. Daripada Anda capek mengetik tombol "Kirim Email" ke 1.000 pelanggan satu per satu, kita suruh komputer memutar roda tugasnya sebanyak 1.000 kali secara otomatis.
* **Untuk Pro:** Di Python, indentasi (spasi/lekukan ke dalam) itu bukan sekadar estetika, tapi **kewajiban sintaks**. Python tidak memakai kurung kurawal `{}` seperti Java atau C++ untuk menandai blok kode, melainkan menggunakan indentasi.

🎨 **Visualisasi Konsep**
```text

### 5.1. Manipulasi Data (Jurus 1: Filtering Ekstrem)

📖 **Konsep**
Di sini kita akan memotong dan menyaring data. Terkadang bos meminta syarat yang rumit, misalnya: *"Carikan penumpang pria, yang umurnya di atas 40 tahun, DAN ada di kelas 1!"*

⚠️ **Peringatan Penting (Sering Error!):**
Di Pandas, kita tidak bisa memakai kata `and` atau `or`.
* Untuk syarat DAN, gunakan simbol **`&`**.
* Untuk syarat ATAU, gunakan simbol garis lurus **`|`** (biasanya di atas tombol Enter).
* **WAJIB** membungkus setiap syarat dengan kurung bulat `()`.

````
# Mencari: Pria (Male) DAN Umur di atas 40
pria_tua = df[ (df['Sex']=='male') & (df['Age'] > 40) ]
display(pria_tua.tail(3))

# Mencari: Penumpang Kelas 1 ATAU Penumpang Kelas 2
# Penumpang kelas 3 tidak akan masuk daftar
kelas_vip = df[ (df['Pclass']==1) | (df['Pclass'] == 2) ]
display(kelas_vip.head(3))
````

## 5.3. Manipulasi Data (Jurus 3: Grouping & Sorting)

📖 **Konsep**
* **`.groupby()`:** Ini adalah jurus memeras data (Pivot Table). Kita mengelompokkan kategori tertentu, lalu menghitung total atau rata-ratanya. Di sinilah *insight* bisnis biasanya ditemukan!
* **`.sort_values()`:** Setelah data diperas, kita bisa mengurutkannya dari yang terbesar ke terkecil (Descending) agar lebih enak dibaca bos.

````
# GROUPING & SORTING
# Pertanyaan Bisnis: Kelas tiket mana yang paling banyak selamat?
# Kita grup berdasarkan 'Pclass', lalu cari rata-rata (.mean) dari kolom 'Survived'.
tabel_keselamatan = df.groupby('Pclass')['Survived'].mean()

# Kita urutkan dari persentase terbesar ke terkecil (ascending=False)
tabel_keselamatan_berurut = tabel_keselamatan.sort_values(ascending=False)

print("Persentase keselamatan berdasarkan kelas tiket (%): ")
print(tabel_keselamatan_berurut)

# Insight: Terbukti! Kelas 1 punya survival rate tertinggi (62%), sedangkan Kelas 3 terendah (24%).

# Latihaan Maul pribadi
````
