<h2 id="var">Tentang Variabel</h2>

Variabel adalah lokasi memori yang dapat digunakan untuk menyimpan suatu nilai dimana nilai tersebut dapat diubah sewaktu-waktu sesuai dengan kebutuhan pengguna. Variabel dalam Python memiliki properti dinamis dan <b>tidak perlu</b> dideklarasikan dengan tipe data tertentu. Variabel python juga dapat diubah saat program dijalankan.

## Indexing dan Slicing pada Python




---




- Pengindeksan dengan Python berarti merujuk ke elemen yang dapat diiterasi berdasarkan posisinya di dalam urutan index.

- Setiap karakter dapat diakses menggunakan nomor indeksnya.

- Untuk mengakses karakter dalam string, terdapat dua cara:
  1. *Positive index number* → menggunakan index nomor positif
  2. *Negative index number* → menggunakan index nomor negatif

 ### *Positive index number* (index bilangan positif)


- Dalam pengindeksan Python Positif, masukkan nilai indeks positif yang ingin diakses menggunakan tanda kurung siku <code>[ ]</code>. Nomor indeks dimulai dari 0 yang menunjukkan karakter pertama dari sebuah string.

image.png

```python
my_str = "WORKSHOP UG"
print(my_str[-4])
```

- Dalam slicing dengan index python positif, masukkan rentang nilai indeks positif yang ingin diakses menggunakan tanda kurung siku <code>[]</code> dan dengan menggunakan simbol titik dua ( **:** ).

Seperti pada contoh di bawah ini:

```python
# Mengambil kata "WORKSHOP"
print(my_str[0:11])
```

`my_str[0:8]` → Menggunakan indeks dari karakter pertama (indeks 0) hingga karakter ke-7 (indeks 7). **Indeks 8 dikecualikan**.

### *Negative index number* (index bilangan negatif)

Dalam pengindeksan negatif dengan Python, masukkan nilai indeks negatif yang ingin diakses menggunakan tanda kurung siku <code>[ ]</code>. Di sini, nomor indeks dimulai dari nomor indeks -1 yang menunjukkan karakter terakhir dari sebuah string.

image.png

```python
my_str = "WORKSHOP UG"
print(my_str[-1:-5])
```

#### Indexing Python List

Pengindeksan Python List dapat dilakukan dengan mengakses nilai <code>list</code> dengan mengacu pada nomor indeksnya. Kali ini akan menggunakan perintah "**(list[2])**" dan mengembalikan **"Kuda"**.

image.png

```python
# Positive index number
list = ['Anjing', 'Kucing', 'Kuda', 'Beruang']
print(list[2])
```

```python
# Negative index number
list = ['Anjing', 'Kucing', 'Kuda', 'Beruang']
print(list[-3])
```

<h3 id="aritmatika">1. Operator Aritmatika</h3>

Operator ini merupakan operator yang biasa digunakan dalam operasi matematika. Daftar operator dapat dilihat pada tabel di bawah ini:

| Operator | Operation |
| -------- | --------- |
|     +    | Penjumlahan |
|     -    | Pengurangan |
|     *    | Perkalian |
|     /    | Pembagian |
|     %    | Modulo atau sisa hasil bagi|
|    **    | Pangkat |
|    //    | Pembagian bulat|

```python
# Contoh Kode Program

# Penjumlahan
a = 10 + 3
print('Hasil penjumlahan 10 + 3 = ', a)

# Pengurangan
b= 10 - 3
print ('Hasil pengurangan 10 - 3 = ', b)

# Perkalian
c = 10 * 3
print('Hasil perkalian 10 x 3 = ', c)

# Pembagian
d = 10 / 3
print('Hasil pembagian 10 : 3 = ', d)

# Modulo
e = 10 % 3
print('Hasil sisa bagi 10 % 3 = ', e)

# Pangkat
f = 10 ** 3
print('Hasil pangkat 10^3 = ', f)

# Pembagian bilangan bulat
g = 10 // 3
print('Hasil pembagian bilangan bulat 10 //3  = ', g)
```

### 3. Operator Perbandingan</h3>

Operator perbandingan atau relasional, yaitu operator yang digunakan untuk melakukan perbandingan antara dua operand dan menghasilkan nilai <code>True</code> atau <code>False</code>.

| Operator | Fungsi | Contoh |
| -------- | -------- | ------- |
|    ==    | Sama dengan   | A == B -> <code>True</code> jika A sama dengan B|
| != or <> | Tidak sama dengan | A != B or A <> B -> <code>True</code> jika A tidak sama dengan B |
|     <    | Lower than | A < B -> <code>True</code> jika A lebih rendah dari B |
|     >    | Greater than | A > B -> <code>True</code> jika A lebih besar dari B |
|    <=    | Lower than or equals | A <= B <code>True</code>jika A lebih rendah atau sama dengan B |
|    >=    | Greather than or equals | A >= B <code>True</code> jika A lebih besar atau sama dengan B |


```python
x = 10
y = 5
z = 2

# > Lebih besar dari
print(x > y)  # True
print(x > y > z)  # True
```

```python
# < Lebih rendah dari / kurang dari
print(x < y)  # False
print(y < x)  # True
```

```python
# Sama dengan
print(x == y)  # False
print(x == 10)  # True
```

```python
# != Tidak sama dengan
print(x != y)  # True
print(10 != x)  # False
```

```python
# >= Lebih besar sama dengan dari
print(x >= y)  # True
print(10 >= x)  # True
```

```python
# <= Lebih rendah sama dengan dari
print(x <= y)  # False
print(10 <= x)  # True
```
